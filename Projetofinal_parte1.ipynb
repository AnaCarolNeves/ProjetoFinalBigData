```python
import pandas as pd
import numpy as np
```

```python
df_cisp = pd.read_csv('BaseDPEvolucaoMensalCisp.csv', sep= ';', encoding = 'latin-1')
df_cisp.head()
```

```python
for i in df_cisp.columns:
    print(i)
```

```python
df_total_furtos = df_cisp[['munic','total_furtos']].groupby('munic').sum().reset_index()
df_total_furtos = df_total_furtos.sort_values(by='total_furtos', ascending=False)
df_total_furtos
```

```python
df_furto_veiculos = df_cisp[['munic','furto_veiculos']].groupby('munic').sum().reset_index()
df_furto_veiculos = df_furto_veiculos.sort_values(by='furto_veiculos', ascending=False)
df_furto_veiculos
```

```python
df_furto_transeunte = df_cisp[['munic','furto_transeunte']].groupby('munic').sum().reset_index()
df_furto_transeunte = df_furto_transeunte.sort_values(by='furto_transeunte', ascending=False)
df_furto_transeunte
```

```python
df_furto_coletivo = df_cisp[['munic','furto_coletivo']].groupby('munic').sum().reset_index()
df_furto_coletivo = df_furto_coletivo.sort_values(by='furto_coletivo', ascending=False)
df_furto_coletivo
```

```python
df_furto_celular = df_cisp[['munic','furto_celular']].groupby('munic').sum().reset_index()
df_furto_celular = df_furto_celular.sort_values(by='furto_celular', ascending=False)
df_furto_celular
```

```python

df_furto_bicicleta = df_cisp[['munic','furto_bicicleta']].groupby('munic').sum().reset_index()
df_furto_bicicleta = df_furto_bicicleta.sort_values(by='furto_bicicleta', ascending=False)
df_furto_bicicleta
```

```python
array_furto_veiculos = np.array(df_furto_veiculos['furto_veiculos'])

maximo = np.max(array_furto_veiculos)
minimo = np.min(array_furto_veiculos)
media = np.mean(array_furto_veiculos)
mediana = np.median(array_furto_veiculos)
distancia = (media-mediana)/mediana
q1 = np.quantile(array_furto_veiculos, 0.25)
q3 = np.quantile(array_furto_veiculos, 0.75)
iqr = q3 - q1
limite_inferior = q1 - (1.5 * iqr)
limite_superior = q3 + (1.5 * iqr)
print(f'Mínimo: {minimo}')
print(f'Limite Inferior: {limite_inferior}')
print(f'Média: {media}')
print(f'Mediana: {mediana}')
print(f'Distância: {round(distancia * 100 , 2)}%')
print(f'Q1: {q1}')
print(f'Q3: {q3}')
print(f'IQR: {iqr}')
print(f'Máximo: {maximo}')
print(f'Limite Superior: {limite_superior}')
```

## 1. Distância <= |10%|: Indica que a média tende a ser uma medida de tendência central confiável
## 2. Distância >= |10%| e Distância < |25%|: Indica que a média está sofrendo moderadamente a influência dos valores extremos da distribuição dos dados e, com isso, pode ser que seja (ou não) uma medida de tendência central confiável(precisará analisar outras medidas que no futuro serão apresentadas)
## 3. Distância >=|25%|: Indica que a média está sofrendo alta a influência dos valores extremos da distribuição dos dados e,com isso, tende a não ser uma medida de tendência central confiável (precisará analisar outra medidas que no futuro serão apresentadas)
